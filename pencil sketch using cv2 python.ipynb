```python
import cv2
import matplotlib.pyplot as plt
import sys
img=cv2.imread("Desktop/phtot of me/queens_land_profile_pic.jpg")
grayscaleimg=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
grayscaleimginv=255-grayscaleimg
grayscaleimginv=cv2.GaussianBlur(grayscaleimginv,(25,25),0)
op=cv2.divide(grayscaleimg,255-grayscaleimginv,scale=256.0)
```

```python
plt.imshow(op,cmap='gray')
```

```python
cv2.namedWindow('img',cv2.WINDOW_AUTOSIZE)
cv2.namedWindow('pencil sketch',cv2.WINDOW_AUTOSIZE)
cv2.imshow('image',img)
cv2.imshow("pencilsketch",op)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
